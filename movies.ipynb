```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split #to split and train the data
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier #for K nearest neighbor
from sklearn.metrics import confusion_matrix # checking the accuracy of the model
from sklearn.metrics import accuracy_score # checking the accuracy of the model
from sklearn.metrics import f1_score #checking the accuracy of the model
```

```python
 data = pd.read_csv(r"C:\Users\Sarah\Desktop\sem 2\biz intelligence\group work\movies data.csv")
```

```python
data
```

```python
non_zero = ['First Release Date','No.Seasons','Duration','Ratings IMBD','Last Release Date','Popularity (0-10)']
for coloumn in non_zero:
    data[coloumn] = data[coloumn].replace(0,np.NaN)
    mean = int(data[coloumn].mean(skipna = True))
    data[coloumn] = data[coloumn].replace(np.NaN, mean)
    print(data[coloumn])
```

```python
import seaborn as sns
p=sns.pairplot(data, hue = 'Outcome')
```

```python
# we are train and testing the data
X =data.iloc[:,0:8]
y =data.iloc[:,8]
X_train,X_test,y_train, y_test = train_test_split(X, y, test_size=4,random_state=0, stratify=y)
```

```python
import math
math.sqrt(len(y_test))
math.sqrt(len(X_test))
```

```python
classifier = KNeighborsClassifier(n_neighbors=13,p=2,metric='euclidean')
classifier.fit(X_train,y_train)
```

```python
y_pred =  classifier.predict(X_test)
y_pred
```

```python
cm= confusion_matrix(y_test,y_pred)
cm
```

```python
print(f1_score(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5, 7))
ax = sns.distplot(data['Outcome'], hist=False, color="r", label="Actual Value")
sns.distplot(y_pred, hist=False, color="b", label="Predicted Values", ax=ax)
plt.title('Actual vs Precited value for outcome')
plt.show()
plt.close()
```
